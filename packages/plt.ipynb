```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [-3,5,1,3,-2,4,-1,0,6]

plt.plot(x,y, marker="o", color="red")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = ["red", "green", "blue", "pink", "black", "grey"]
y = [4, 1, 8, 3, 12, 5]

plt.bar(x,y, color=["red", "green", "blue", "pink", "black", "grey"])
plt.title("Colors")
plt.xlabel("Colors")
plt.ylabel("Count")
plt.show()
```

```python
x = ["red", "green", "blue", "pink", "black", "grey"]
y = [4, 1, 8, 3, 12, 5]

plt.pie(y, labels=x, autopct="%1.1f%%", colors=["red", "green", "blue", "pink", "black", "grey"])
plt.title("Colors")
plt.xlabel("Colors")
plt.ylabel("Count")
plt.show()
```
